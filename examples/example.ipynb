```python
from more_itertools import last
import matplotlib.pyplot as plt
import patchyAnalysisTools.trajectory as trj
```

```python
# load trajectory
traj = trj.read_trajectory("trajectory.conf")

# get last frame
last_frame = traj.get_last_frame()

# calculate g(r)
r,gr = last_frame.calculate_rdf()
```

```python
# plot radial distribution function
plt.plot(r*50,gr)
plt.xlabel('r (A)')
plt.ylabel('g(r)')
plt.title('radial distribution function')
plt.show()
```

```python
# calculate and plot s(q)
k,sk = last_frame.calculate_sq()
plt.loglog(k/50,sk)
plt.xlabel('q (A^{-1}')
plt.ylabel('s(q)')
plt.show()
```
